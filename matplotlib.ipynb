```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

```python
#plotting the data
```

```python
#sine wave
plt.plot(x,y)
plt.show()
```

```python
#cosine wave
plt.plot(x,z)
plt.show()
```

```python
#adding title x-axis , y-axis and y-axis labels
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('sine wave')
plt.show()
```

```python
#parabola
x = np.linspace(-10,10,20)
y = x**2
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('parabola')
plt.show()
```

```python
plt.plot(x,y,"r+")
plt.xlabel('x')
plt.ylabel('y')
plt.title('parabola')
plt.show()
#
```

```python
plt.plot(x,y,"g*")
plt.xlabel("x")
plt.ylabel("y")
plt.title("parabola")
plt.show()
```

```python
x = np.linspace(-5,5,50)
plt.plot(x,np.sin(x),"g-")
plt.plot(x,np.cos(x),"r--")
plt.show()
```

```python
#bar plot
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
langs = ['C','C++','Java','Python','PHP']
students = [23,17,35,29,12]
ax.bar(langs,students)
plt.xlabel("languages")
plt.ylabel("number of students")
plt.title("students enrolled")
plt.show()
```

```python
#piechart
fig1 = plt.figure()
ax = fig1.add_axes([0,0,1,1])
langs = ['C','C++','Java','Python','PHP']
students = [23,17,35,29,12]
ax.pie(students,labels = langs, autopct = "%1.1f%%")
plt.show()
```

```python
#Scatter plot
x = np.linspace(0,10,30)
y = np.sin(x)
plt.scatter(x,y)
plt.show()
```

```python
x = np.linspace(0,10,30)
y = np.sin(x)
z = np.cos(x)
plt.scatter(x,z,color = "g")
plt.scatter(x,y,color = "r")
plt.show()
```

```python
x = np.linspace(0,10,30)
y = np.sin(x)
z = np.cos(x)
plt.scatter(x,z,color = "g",label = "sine")
plt.scatter(x,y,color = "r",label = "cosine")
plt.legend()
plt.show()
```

```python
#3d scatter plot
fig3 = plt.figure()
ax = plt.axes(projection = "3d")
z = 20 * np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter3D(x,y,z,c = z,cmap = "Greens")
plt.show()
```
